```python
# Import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import f_oneway
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

```

### Question 01:   How many sales per month in the year (2013 & 2014) do we have?

```python


# her I make a new column by name Date and Convert 'Date' column to datetime type
dataset['Date'] = pd.to_datetime(dataset['Date'])

# Filter sales data for the years 2013 and 2014
salesByMonth = dataset[(dataset['Year'] == 2013) | (dataset['Year'] == 2014)]

# Group by month and count the number of sales
sales_per_month = salesByMonth.groupby(['Year', 'Month']).size().reset_index(name='Sales Count')

print("Sales per month in the years 2013 & 2014:\n")
print(sales_per_month)
```

### Question 02:   Which country has the most sales quantity of sales?

```python

# Group by country and sum the sales quantity
sales_by_country = dataset.groupby('Country')['Order_Quantity'].sum().reset_index()

# Find the country with the maximum sales quantity
country_with_most_sales = sales_by_country.loc[sales_by_country['Order_Quantity'].idxmax()]

print("Country with the most sales quantity is:", country_with_most_sales['Country'])
print("With Total sales quantity:", country_with_most_sales['Order_Quantity'])

```

### Question 03: How many orders were made in each region (state) of France?

### Question 04. Get the top-5 sales with the highest revenue

### Question 05. Are all the age groups providing equal average profit? apply the appropriate inferential analysis and interpret the results.

```python
# Perform ANOVA to test if there is a significant difference in profit among age groups
# Assuming 'Age_Group' is the column containing age group information and 'Profit' is the column containing profit
groups = []
for age_group, group_data in dataset.groupby('Age_Group'):
    groups.append(group_data['Profit'])

# Perform one-way ANOVA
statistic, p_value = f_oneway(*groups)

# Set significance level
alpha = 0.05

print("Results of one-way ANOVA:")
print("F-statistic:", statistic)
print("p-value:", p_value)

if p_value < alpha:
    print("Reject the null hypothesis. There is significant evidence that at least one age group's mean profit is different.")
else:
    print("Fail to reject the null hypothesis. There is no significant evidence that the mean profit differs among age groups.\n")
    
print(f'Profit = -22185293586238.848 + 30261039228373.39 * {dataset.Order_Quantity} +  30261039228383.4 * {dataset.Age_Group} + 30261039228358.02 * {dataset.Product_Category} +  128265.85807291396')
```

### Question 06. Use 80% of the data, to build a regression model for predicting the profit based on features such as order quantity, age group, and product category.

```python
# Define features and target variable
features = ['Order_Quantity', 'Age_Group', 'Product_Category']
target = 'Profit'

# Split data into training and testing sets
train_data, test_data = train_test_split(dataset, test_size=0.2, random_state=42)

# Define preprocessing steps for categorical variables
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), ['Age_Group', 'Product_Category'])
    ],
    remainder='passthrough'
)

# Define the regression model
model = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

# Train the regression model
model.fit(train_data[features], train_data[target])

# Print intercept and coefficients
intercept = model.named_steps['regressor'].intercept_
coefficients = model.named_steps['regressor'].coef_

print("Intercept:", intercept)
print("Coefficients:")
for feature, coef in zip(features, coefficients):
    print(f"{feature}: {coef}")

# Evaluate the model
predictions = model.predict(test_data[features])
mse = mean_squared_error(test_data[target], predictions)
print("Mean Squared Error:", mse)
```

### Question 07. Use the rest of the data set to report the performance (R2, RMSE, etc.) of the regression model for predicting profit.

```python


# Use the rest of the data set (the data not used for training) for evaluation
eval_data = dataset.drop(train_data.index)

# Predict profit using the regression model
predictions = model.predict(eval_data[features])

# Calculate R-squared
r_squared = r2_score(eval_data[target], predictions)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(eval_data[target], predictions))

print("Evaluation Metrics:")
print("R-squared:", r_squared)
print("Root Mean Squared Error (RMSE):", rmse)
```
